# Processing image

```python
img_size= 224
def preprocess_image(filepath):
    img= tf.io.read_file(filepath)
    img= tf.image.decode_jpeg(img, channels= 3)
    img= tf.image.convert_image_dtype(img, tf.float32)
    img= tf.image.resize(img, size=[img_size,img_size])

    return img
```

```python
rr= tf.io.read_file(mountain[12])
rr=tf.image.decode_jpeg(rr,channels=3)
rr=tf.image.convert_image_dtype(rr,tf.float32)
rr= tf.image.resize(rr,size=[224,224])
rr
```

```python
def get_label(filepath, label):
    img= preprocess_image(filepath)

    return img, label
```

# Create batches

```python
Batches =32
def create_batches(x,y=None, test_data=False,validation_data=False,Batches=Batches):
    if test_data:
        print('Creating testing data batches...')
        data= tf.data.Dataset.from_tensor_slices((tf.constant(x)))
        data_batch= data.map(preprocess_image).batch(Batches)
        return data_batch

    elif validation_data:
        print('Creating validation data batches...')
        data= tf.data.Dataset.from_tensor_slices((tf.constant(x),
                                                  tf.constant(y)))
        data_batch= data.map(get_label).batch(Batches)
        return data_batch
    else:
        print("Creatting traing data batches...")
        data= tf.data.Dataset.from_tensor_slices((tf.constant(x),
                                                  tf.constant(y)))
        data=data.shuffle(buffer_size=len(x))
        data_batch= data.map(get_label)
        data_batch= data_batch.batch(Batches)
        return data_batch
```

```python
train_data= create_batches(x_train[:500],y_train[:500])
```

```python
val_data = create_batches(x_test[:500],y_test[:500],validation_data=True)
```

```python
train_data.element_spec
```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Define input shape
input_shape = (224, 224, 3)

# Create a Sequential model
model = Sequential()

# Add input layer
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
# Add convolutional layer
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))

# Finish with output layer
model.add(Dense(5, activation='softmax')) 


# Compile model
model.compile(
        loss= tf.keras.losses.CategoricalCrossentropy(),
        optimizer = tf.keras.optimizers.Adam(),
        metrics= ['accuracy']
    )
# Print model summary
model.summary()
```

```python
early_stop = tf.keras.callbacks.EarlyStopping(monitor='accuracy', patience=1)
```

```python
model.fit(train_data,
    validation_data= val_data,
    validation_freq=1,
    epochs=20,
    callbacks= early_stop
)
```
